#### Task 4.Data Cleaning

```python
# dropping all missing values rows
df.dropna(inplace = True)
df.isnull().sum()
```

```python
# dealing with duplicate rows
df.duplicated().sum()     

# deleting all duplicates rows
df.drop_duplicates(inplace = True)       # 12 duplicate rows deleted
df.duplicated().sum()
```

```python
df.dtypes  # id,host_id dtype(int)

# changing data type of id ,host_id (object)
df['id'] = df['id'].astype(object)
df.dtypes

df['host_id'] = df['host_id'].astype(object)
df.dtypes
```

## Exploratory Data Analysis
#### Task 5.Data Analysis

#### Univariate Analysis

```python
# identifying Outliers in price
df2 = df[df['price']<1500]     # as there are outliers above 1500

plt.figure(figsize=(4,3))
sns.boxplot(data = df2 ,x='price')
plt.show()  # to show only graphs
```

```python
# Price Distribution after removing Outliers

plt.figure(figsize=(5,4))
plt.ylabel("Frequency")
sns.histplot(data=df2, x='price',bins=100) 
plt.title("Price Distribution")
plt.show()
```

```python
# for only float dtypes (number columns), histogram will be created
df.columns
```

```python
# for how many days each of the properties were available


plt.figure(figsize=(5,3))
sns.histplot(data=df, x='availability_365') 
plt.title("Availability_365 Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
df.dtypes
```

```python
# average price according to area

df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
# average price per bed after doing feature engineering (adding a new column)

df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

### Feature Engineering

```python
# adding a column ['price per bed']

df['price per bed']=df['price']/df['beds']
df.head()
```

#### Bivariate Analysis

```python
# price dependencies on neighbourhood
sns.barplot(data=df,x= 'neighbourhood_group',y='price',hue='room_type')
plt.show()
```

```python
# number of reviews and price relationship

plt.figure(figsize=(8,5))
plt.title("Locality and review dependencies")
sns.scatterplot(data=df2, x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
# pairplots for various plots and many variables

sns.pairplot(data = df2,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
plt.show()
```

```python
# Geographical Distribution of AirBnb listing

sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.title('Geographical Distribution of AirBnb listing')
plt.show()
```

```python
# Heatmap - Correlation of one variable with others for numerical column
# It is only used for number column, here we are going to use all number columns

corr = df2[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(data=corr,annot=True)
plt.title('Correlation Matrix')
plt.show()
```
